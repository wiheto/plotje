# Plotje examples

This is a quick notbook to show how plotje works.

It is mainly a helper function to style figures the same way with minimal effort. Based on some of the ideas of making graphs by Edward Tufte (e.g. less black in figures, removing graph clutter)

Some matplotlib functions are not included in plotje, most noteable: set_xticks, set_yticks, and tight_layout. Setting ticks should be done before. Tight_layout should be done after calling plotje.

```python
import plotje
import numpy as np 
import matplotlib.pyplot as plt 
```

## Difference in style with default matplotlib

First let's make some data.

```python
x = np.random.normal(0,1,30)
y = np.random.normal(0,1,30)
z = np.random.normal(0,1,30)

```

Now lets look at before and after with plotje

```python
fig, ax = plt.subplots(1,2)
# Before
ax[0].scatter(x,y)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('Default Matplotlib')

# Plotje
ax[1].scatter(x,y)
plotje.styler(ax[1], xlabel='x', ylabel='y', title='With Plotje')
plt.tight_layout()
```

## Parameter dictionaries

Allowing for one dictionary that can configure most of the figure is very helpful instead of repeating the same lines of code again and again. 

In the example below, all general parameters are set in a dictionary "params" that are used across all subplots.

```python
params = {
    'xlabel': 'x', 
    'fontweight': 'bold', 
    'aspectsquare': True
}

fig,ax = plt.subplots(1,2)
ax[0].scatter(x,y)
ax[1].scatter(x,z)

ylabs = ['y', 'z']
for i, a in enumerate(ax): 
    plotje.styler(a, ylabel=ylabs[i], **params)

plt.tight_layout()
```

# Easy figure legends

Figure legend also become easy to use

```python
params = {
    'xlabel': 'x',
    'aspectsquare': True, 
    'legend': True, 
    'title': 'Legend example'
}

fig,ax = plt.subplots()
ax.scatter(x,y, color='Red', label='XY')
ax.scatter(x,z, color='Blue', label='YZ')

plotje.styler(ax, ylabel=ylabs[i], **params)

plt.tight_layout()


```

```python
"""
# Customizeable axes

It is easy to turn off one of the axis. It is also possible to move the offset of the shown axis. 
"""
```

```python
fig,ax = plt.subplots(1)
ax.hist(x)
plotje.styler(ax, leftaxis='off', xoffset=-0.025, title='Example Histogram')
```

```python
"""
# Easy colorbars

Colorbars are also easier to control
"""
```

```python
data = np.random.normal(0,1,(10,10))
fig,ax = plt.subplots(1)
himg = ax.imshow(data, cmap='RdBu_r')
plotje.styler(ax, colorbar=himg, colorbarlabel='data', cbaroffset=0.5, title='imshow example')


```
